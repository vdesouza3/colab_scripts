```python
import wradlib as wrl
import matplotlib.pyplot as plt # biblioteca para plotar mapas
import numpy as np              # biblioteca para trabalhar com matrizes
```

###**3) Plota Estratégia**

```python
nrays = 360                                                                   # azimute = ângulo entre o feixe do radar e o Norte
nbins = 1000                                                                  # bin = a menor info que tem do radar. Quanto mais bins, mais detalhes.
range_res = 250                                                               # resolução radial = definida pelo tamanho do bin em metros
ranges = np.arange(nbins) * range_res                                         # alcance do radar
elevs = [1.8] # elevações = ângulos entre o feixe do radar e o plano horizontal (em graus)
sitecoords = (-5.904480, -35.254010, 78.0)                                    # latitude, longitude e altitude do radar
beamwidth = 0.972                                                             # largura do feixe

ax = wrl.vis.plot_scan_strategy(ranges,elevs,sitecoords,
                                beamwidth = 0.972,
                                vert_res=1000.0,                              # resolução vertical = 1000 m = 1 km
                                maxalt=15000.0,                               # altura máxima
                                range_res=50000.0,
                                maxrange=250000.0,                            # alcance máximo = 250000 m = 250 km
                                units="km")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import wradlib as wrl

# Parâmetros gerais
nrays = 360
nbins = 1000
range_res = 250
ranges = np.arange(nbins) * range_res
beamwidth = 0.972

# Dados para os locais
locais = [
    {"nome": "Natal", "elevs": [20.0, 16.7, 13.9, 11.4, 9.4, 7.6, 6.0, 4.7, 3.6, 2.6, 1.8, 1.1, 0.5], "coords": (-5.904480, -35.254010, 78.0)},
    {"nome": "Petrolina", "elevs": [32.0, 24.6, 20.3, 16.7, 13.6, 10.9, 8.6, 6.6, 5.0, 3.5, 2.3, 1.2, 0.3], "coords": (-9.367, -40.573, 394.0)},
    {"nome": "Maceió", "elevs": [25.0, 21.7, 18.7, 16.1, 13.8, 11.8, 10.0, 8.5, 7.1, 5.9, 4.8, 3.8, 3.0, 2.2, 1.6, 1.0], "coords": (-9.551290, -35.770680, 102.0)},
    {"nome": "Salvador", "elevs": [25.0, 21.8, 18.9, 16.4, 14.2, 12.1, 10.4, 8.8, 7.4, 6.1, 5.0, 4.0, 3.1, 2.4, 1.7, 1.0, 0.5], "coords": (-12.9, -38.32, 30.0)}
]

# Criar figuras separadas para cada local
for local in locais:
    fig, ax = plt.subplots(figsize=(8, 4))

    ax = wrl.vis.plot_scan_strategy(
        ranges,
        local["elevs"],
        local["coords"],
        beamwidth = 0.972,
        vert_res=1000.0,
        maxalt=15000.0,
        range_res=50000.0,
        maxrange=250000.0,
        units="km",
        terrain=None,
        ax=plt.gca()  # Obter o eixo atual com plt.gca()
    )

    coords = local["coords"]
    latitude, longitude, altitude = coords
    ax.set_title(f"Estratégia de Varredura - {local['nome']}\nLat: {latitude}, Lon: {longitude}, Alt: {altitude} m")

    # Criar uma legenda apenas para os valores das elevações
    elevs_legend = [f"{elev}°" for elev in local["elevs"]]

    # Definir manualmente a legenda no canto superior direito
    ax.annotate(f'Elevações: {", ".join(elevs_legend)}', xy=(1, 1), xycoords='axes fraction', fontsize=8, ha='right', va='top')

    # Exibir a figura
    plt.show()
```
